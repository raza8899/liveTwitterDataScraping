### Data Scraping through Tweepy

```python
###Python scraping
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream

access_token = "yourAccesToken"
access_token_secret= "access_token_secret"
consumer_key = "consumer_key"
consumer_secret = "consumer_s"


class StdOutListener(StreamListener):
    def on_data(self, data):
        try:
            with open('umairBhai.json', 'a') as f:
                f.write(data)
                return True
        except BaseException as e:
            print("error on data: %s" % str(e))
        return True
    def on_error(self, status):
        print (status)

if __name__=='__main__':
    l = StdOutListener()
    auth = OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    stream = Stream(auth, l)
    stream.filter(track=['#AACR20'])
```

### Converting the scraped data into the csv file

```python
import json

data = []
with open('bjp.json') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
data[19]
```

```python
my_dict=[]
index_list=[]
name_list=[]
date_list=[]
val=1
for x in data:
    print(str(val)+'\n-----')
    if 'retweeted_status' not in x:
        if 'extended_tweet' not in x:
            y=x['text']
            print(y)
        else:
            y=x['extended_tweet']['full_text']
            print(y)
    else:
        if 'extended_tweet' not in x['retweeted_status']:
            y=x['retweeted_status']['text']
            print(y)
        else:
            y=x['retweeted_status']['extended_tweet']['full_text']
            print(y)
    print('\n\n\n')
    my_dict.append(y)
    index_list.append(val)
    name_list.append(x['user']['name'])
    date_list.append(x['created_at'])
    val+=1
```

```python
import pandas as pd
columns=["Index","Date","Name","Tweet"]
my_df=pd.DataFrame(columns=columns)
```

```python
my_df['Index']=index_list
my_df['Date']=date_list
my_df['Name']=name_list
my_df['Tweet']=my_dict
```

```python
my_df.to_csv('umair.csv', encoding='utf-8', index=False)
```
