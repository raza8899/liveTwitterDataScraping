### Readind the Scraped Data and filtering the needed things

```python
import pandas as pd
import json

data = []
with open('dataScraped.json') as f:
    for line in f:
        data.append(json.loads(line))
```

#### A look at raw Json Data


```python
data[0]
```

### Getting the fullText from the Tweets and other entities like Name, Date etc.

```python
my_dict=[]
index_list=[]
name_list=[]
date_list=[]
val=1
for x in data:
    if 'retweeted_status' not in x:
        if 'extended_tweet' not in x:
            y=x['text']
            
        else:
            y=x['extended_tweet']['full_text']
            
    else:
        if 'extended_tweet' not in x['retweeted_status']:
            y=x['retweeted_status']['text']
            
        else:
            y=x['retweeted_status']['extended_tweet']['full_text']
            
    
    my_dict.append(y)
    index_list.append(val)
    name_list.append(x['user']['name'])
    date_list.append(x['created_at'])
    val+=1
```

```python
import pandas as pd
columns=["Index","Date","Name","Tweet"]
my_df=pd.DataFrame(columns=columns)
my_df['Index']=index_list
my_df['Date']=date_list
my_df['Name']=name_list
my_df['Tweet']=my_dict
```

### Finally exporting the data as excel or csv file for further Analysis

```python
my_df.to_excel('scrapedData.xlsx', index=False)
```

```python
my_df.to_csv('scrapedData.csv',index=False)
```
